```python
import numpy as np
import matplotlib.pyplot as plt

cgpa = np.array([6,6.5,7,7.5,8,8.9,9.2,9.5,10])
package = np.array([4,4.5,5,5.8,6,6.2,6.8,7,7.5,8,8.8])
font1 =  {'family':'serif','color':'blue','size':15}
font2 =  {'family':'serif','color':'darkred','size':20}
plt.title("CGPA VS PACKAGE",fontdict=font1)
# Truncate the longer array to match the length of the shorter array
package = package[:len(cgpa)]
plt.plot(cgpa,package,"d:r",ms=15,mec="orange",mfc="black",
         linewidth=2.4) #ms=marker size mfc =marker face color,mec =marker edge color
plt.xlabel("CGPA",fontdict=font1)
plt.ylabel("PACKAGE",fontdict=font1)

plt.title("CGPA VS PACKAGE",fontdict = font2)
plt.show()
```

```python
#bar plot --->it works on numerical as well as categorial data
fruit = np.array(["apple","banana","mango","watermelon"])
quantity = np.array([3,4,5,6])
color = np.array(["red","yellow","yellow","green"])
plt.bar(fruit,quantity,color=color)
plt.show()
```

```python
fruit = np.array(["apple","banana","mango","watermelon"])
quantity = np.array([3,4,5,6])
plt.barh(fruit,quantity)
plt.show()
```

```python
#pie plot--> this plot works on numerical as well as categorical data
fruit = np.array(["apple","banana","mango","watermelon"])
percentage_of_fruit = np.array([25,20,25,30])
plt.pie(percentage_of_fruit,labels=fruit)
plt.legend(title="Fruit")
plt.show()
```

```python
#scatter plot --> it works on numerical data ,data represent in the form of dot .
age_of_car = np.array([4,5,6,7,8])
speed_of_car = np.array([150,140,135,128,125])
plt.scatter(age_of_car,speed_of_car)
plt.show()
```
